```python
# Bibliotecas
from pathlib import Path
import sys
import joblib

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.features.preprocessamento_rf import carregar_dados
```

```python
# Fun??o principal
def main():
    arquivo_entrada = PROJECT_ROOT / 'data' / 'raw' / 'desafio_nps_fase_1.csv'
    arquivo_modelo = PROJECT_ROOT / 'models' / 'modelo_rf.pkl'
    arquivo_saida = PROJECT_ROOT / 'outputs' / 'data_classificacao.csv'
    arquivo_saida.parent.mkdir(parents=True, exist_ok=True)

    df, X_pre, preprocesso = carregar_dados(arquivo_entrada)
    modelo = joblib.load(arquivo_modelo)

    df['nps_classificacao_prevista'] = modelo.predict(X_pre)
    df.to_csv(arquivo_saida, index=False, encoding='utf-8-sig')

    print(f'Arquivo salvo com sucesso: {arquivo_saida}')
    print(df[['nps_classification', 'nps_classificacao_prevista']].head())
```

```python
# Executa o programa
if __name__ == '__main__':
    main()
```
